```python
# Importar librerías
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_openai import OpenAIEmbeddings, ChatOpenAI
from langchain_community.vectorstores import Chroma
from langchain_core.prompts import ChatPromptTemplate
from langchain.tools import tool
from langchain_classic.prompts import ChatPromptTemplate
from langchain_classic.schema.runnable import RunnablePassthrough
from langchain_classic.schema.output_parser import StrOutputParser
```

```python
#Cargar el documento pdf de APA 7maedición
file_path = "/content/APA_7ed.pdf"
loader = PyPDFLoader(file_path)
docs = loader.load()
```

```python
# Dividir el documento en chunks
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200,
    length_function=len
)
chunks = text_splitter.split_documents(docs)

print(f"Dividido {len(docs)} documentos en {len(chunks)} chunks.")
```

```python
# Crear el vectorstore
embeddings = OpenAIEmbeddings()
vectorstore = Chroma.from_documents(
    documents=chunks,
    embedding=embeddings,
    persist_directory="./apa_7_vectorstore"
)
```

```python
# Definir el retriever
retriever = vectorstore.as_retriever(search_kwargs={"k": 3})
```

```python
# Configurar el modelo LLM
llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0)
```

```python
# Crear el sistema RAG para corrección APA
def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)
```

```python
# Crear el prompt template para corrección de citas APA
rag_prompt = ChatPromptTemplate.from_template("""
Eres un especialista en normas APA 7ma edición. Tu tarea es corregir citas que están mal escritas según las reglas APA.

CONTEXTO (Manual APA 7ma edición):
{context}

CITA A CORREGIR:
{input}

INSTRUCCIONES:
1. Analiza la cita proporcionada
2. Identifica los errores según las normas APA del contexto
3. Proporciona la cita corregida
4. Explica brevemente qué se corrigió y por qué

Responde en el siguiente formato:
**Cita Original:** [cita original]
**Cita Corregida:** [cita corregida]
**Explicación:** [explicación de las correcciones]

Si la cita ya está correcta, indícalo y explica por qué cumple con las normas APA.
""")
```

```python
# Crear la cadena RAG
rag_chain = (
    {"context": retriever | format_docs, "input": RunnablePassthrough()}
    | rag_prompt
    | llm
    | StrOutputParser()
)
```

```python
# Definir la herramienta de corrección APA
@tool
def apa_citation_corrector(citation: str) -> str:
    """
    Corrige citas APA mal escritas según el manual de APA 7ma edición.

    Args:
        citation (str): La cita que necesita ser corregida

    Returns:
        str: La cita corregida con explicación de los cambios
    """
    try:
        response = rag_chain.invoke(citation)
        return response
    except Exception as e:
        return f"Error al procesar la cita: {str(e)}"
```

```python
# Función auxiliar para extraer citas de texto
@tool
def extract_and_correct_citations(text: str) -> str:
    """
    Extrae todas las citas de un texto y las corrige individualmente según APA 7ma edición.

    Args:
        text (str): Texto que contiene citas a corregir

    Returns:
        str: Reporte con todas las citas corregidas
    """
    # Patrón simple para identificar citas (puedes ajustarlo según tus necesidades)
    citation_pattern = r'\([^)]*[A-Za-z]+[^)]*\)'
    citations = re.findall(citation_pattern, text)

    if not citations:
        return "No se encontraron citas en el texto proporcionado."

    results = []
    for i, citation in enumerate(citations, 1):
        results.append(f"--- CITA {i} ---")
        results.append(f"Original: {citation}")
        correction = apa_citation_corrector(citation)
        results.append(f"Corrección: {correction}")
        results.append("")

    return "\n".join(results)
```

```python
# Si toolkit es una lista
toolkit = []  # Inicializa toolkit como una lista vacía si aún no está definido

toolkit.append("apa_citation_corrector")
toolkit.append("extract_and_correct_citations")

print("✅ Sistema RAG de corrección APA implementado exitosamente!")
print("📚 Herramientas disponibles:")
print("   - apa_citation_corrector: Corrige una cita individual")
print("   - extract_and_correct_citations: Extrae y corrige múltiples citas de un texto")
```
